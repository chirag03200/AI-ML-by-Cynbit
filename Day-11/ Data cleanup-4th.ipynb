```python
import numpy as np
import pandas as pd

df = pd.read_csv("KaggleV2-May-2016.csv")
# df_raw = df.copy()
df.head()
```

```python
# Rename confusing columns
df.rename(columns={"No-show": "NoShow"}, inplace=True)
df
```

```python
# Convert to datetime
df["ScheduledDay"] = pd.to_datetime(df["ScheduledDay"])
df["AppointmentDay"] = pd.to_datetime(df["AppointmentDay"])
df.head()
```

```python
# Fix inconsistent 'NoShow' column (Yes = No Show)
df["NoShow"] = df["NoShow"].map({"Yes": 1, "No": 0})
df.head()
```

```python
# Remove negative age 
df = df[df["Age"] >= 0 & (df["Age"] <= 90)]
df.head()
```

```python
# Drop ID columns
df.drop(columns=["PatientId", "AppointmentID"])
```

```python
# Handle missing values
df.dropna
```

```python
df.head()
```

```python
# Save
df.to_csv("no_show_clean.csv")
```
