```python
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv("trains.csv")
df
```

```python
# Basic overview
print(df.info())
print(df.describe(include='all'))
```

```python
 # Handle missing values
df.drop(columns=["Alley", "PoolQC", "Fence", "MiscFeature"], inplace=True)  # Drop columns with too many NaNs
df.dropna(subset=["LotFrontage"], inplace=True)  # Drop rows with missing frontage
df["GarageType"] = df["GarageType"].fillna("None")  # Fix GarageType safely
df
```

```python
# Fix mixed datatypes
df["MSSubClass"] = df["MSSubClass"].astype(str)  # Categorical not numeric
df
```

```python
# Encode categoricals
df = pd.get_dummies(df, drop_first=True)
df
```

```python
# Remove skewed distributions (example)
df["SalePrice"] = np.log1p(df["SalePrice"])
df
```

```python
# Save cleaned
df.to_csv("house_prices_clean.csv", index=False)
df
```
