## Histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
def calculate_results(marks):
   total = sum(marks)
   average = total/len(marks)

   if average >= 90:
       grade = 'A'
   elif average>= 75:
       grade = 'B'
   else:
       grade = 'C'

   return total, average, grade

# input number of sturdents
stu_num = int(input("Number of students: "))

data = []
# loop for input marks
for i in range(stu_num):
    name = input("Name :")
    print(f"\n enter the marks of student{i+1}: ")
    s1 = float(input("Subject 1 marks: "))
    s2 = float(input("Subject 2 marks: "))
    s3 = float(input("Subject 3 marks: "))
    print("\n")
    total, avg, grade = calculate_results([s1, s2, s3])

    print(f"\nReport for {name}:")
    print(f"Total: {total}, Average: {avg:.2f}, Grade: {grade}")

    #Store the data in a pandas DataFrame
    data.append({ "Name":name, "Subject 1": s1,"Subject 2": s2,"Subject 3": s3,
                 "Total": total, "Average": avg, "Grade": grade})

df = pd.DataFrame(data)
print("\nðŸ“‹ Student Report Card:")
print(df)
```

```python
# Histogram of Total Marks
plt.figure(figsize=(8, 5))
plt.hist(df['Total'], bins=5, color='skyblue', edgecolor='black')
plt.title('Distribution of Total Marks')
plt.xlabel('Total Marks')
plt.ylabel('Number of Students')
plt.grid(True)
plt.show()
```

```python
# Histogram of Average Marks
plt.figure(figsize=(8, 5))
plt.hist(df['Average'], bins=5, color='lightgreen', edgecolor='black')
plt.title('Distribution of Average Marks')
plt.xlabel('Average Marks')
plt.ylabel('Number of Students')
plt.grid(True)
plt.show()
```

## with seaborn

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Total'], bins=5, kde=True, color='skyblue')
plt.title('Distribution of Total Marks')
plt.xlabel('Total Marks')
plt.ylabel('Number of Students')
plt.show()
```

```python
# Histogram of Average Marks using seaborn
plt.figure(figsize=(8, 5))
sns.histplot(df['Average'], bins=5, kde=True, color='lightgreen')
plt.title('Distribution of Average Marks')
plt.xlabel('Average Marks')
plt.ylabel('Number of Students')
plt.show()
```
